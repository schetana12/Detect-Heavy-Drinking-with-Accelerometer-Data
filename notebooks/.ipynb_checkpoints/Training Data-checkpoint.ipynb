```python
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(12,8)})

from collections import Counter
from sklearn.model_selection import train_test_split
```

## Reading TAC DATA

```python
data_dir = "../data/clean_tac/"
list_dir = os.listdir(data_dir)
```

```python
tacs = {}
for pid_csv in list_dir:
    pid = pid_csv[:6]
    tacs[pid] = pd.read_csv(data_dir+pid_csv)
```

## Reading ACCELEROMETER DATA

```python
data_dir = "../data/all_accelerometer_data_pids_13.csv"
acc_data = pd.read_csv(data_dir)
```

## Merge the two datasets

```python
pids = tacs.keys()
pids
```

```python
merge_data['DK3500'].head()
```

```python
all_data = []
for pid in pids:
    df1 = tacs[pid]
#     Shift TAC reading by 45 minutes before merging with accelerometer data
    df1['shift_timestamp'] = df1['timestamp'] - (45*60)
    df2 = acc_data[acc_data['pid']==pid]
#     Change timestamp in accelerometer data from milisec to sec
    df2['timestamp'] = df2['time']//1000 
#     Merge the TAC reading and accelerometer data on timestamp
    tmp = df1.merge(df2, left_on='shift_timestamp', right_on='timestamp')
    tmp['label'] = np.where(tmp['TAC_Reading']<0.08, 0, 1)

    all_data.append(tmp)
large_df = pd.concat(all_data)
```

```python
large_df = large_df.drop('timestamp_y', axis=1)
```

## Split into train, test

```python
X_train, X_test, y_train, y_test = train_test_split(large_df.loc[:, ~large_df.columns.isin(['label','TAC_Reading','pid','time','timestamp_x'])],
                                                    large_df['label'], 
                                                    test_size=0.30, 
                                                    random_state=42)
```

```python
X_train.head()
```

```python
print(f"label distribution in train data: {Counter(y_train)}")
print(f"label distribution in test data {Counter(y_test)}")
```

```python
X_train = X_train.drop('shift_timestamp', axis=1)
X_test = X_test.drop('shift_timestamp', axis=1)
```

## Logistic Regression with only the numeric attributes:

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(fit_intercept = False).fit(X_train, y_train)
y_pred = clf.predict(X_test)
# y_pred = clf.predict_proba(X_test)
print(f'The accuracy on test data is {clf.score(X_test, y_test)}')
```

#### Turns out the regularization parameter defaults to a value that causes an underfit. So I added C=10 to the constructor (C is the inverse so higher value means less regularization). Another change I am making is removing the first feature variable because a bias term is automatically added unless you pass "fit_intercept = False" in the constructor. 

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
